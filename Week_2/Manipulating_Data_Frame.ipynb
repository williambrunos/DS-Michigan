Let's take the name of the presidents and separate then into two columns: 'first' and 'last' refering itselves to the first
and last name of each president, respectvely.

How? Using the ``aply()`` function, which takes as argument a pre built function and the axis which we want to iteraate it over. In this case, will be the ``axis='columns'``

```python
def split_name(row):
    row['First'] = row['President'].split(' ')[0]
    row['Last'] = row['President'].split(' ')[-1]
    return row
```

```python
df = df.apply(split_name, axis='columns')
df
```
