```python
def return_region(x):
    north_east = ('Canada–All Maine Vermont New_Hampshire Massachusetts Rhode_Island New_York Pennsylvania Connecticut New_Jersey Delaware Maryland West_Virginia').split(' ')
    south_east = ('Virginia North_Carolina South_Carolina Georgia Florida Tennessee Alabama Mississippi Arkansas Louisiana').split(' ')
    mid_west = ('Ohio Michigan Indiana Kentucky Wisconsin Illinois Minnesota Iowa Missouri North_Dakota South_Dakota Nebraska Kansas').split(' ')
    south_west = ('Oklahoma Texas').split(' ')
    rocky_moutain = ('Montana Wyoming Idaho Colorado Utah New_Mexico Arizona').split(' ')
    pacific = ('Washington Oregon Nevada California Alaska Hawaii').split(' ')
    southern_california = 'Southern_California'
    
    if x in north_east:
        return "North_East"
    elif x in south_east:
        return "South_East"
    elif x in mid_west:
        return "Mid_West"
    elif x in south_west:
        return "South_West"
    elif x in rocky_moutain:
        return "Rocky_Moutain"
    elif x in pacific:
        return "Pacific"
    elif x in southern_california:
        return "Southern_California"
```

# "Apply" Function on Pandas:

This functions applys a external function to each cell of a row or a column specified.

```python
# Making a list of cell values on "Home State" column cleaned -> Replacing " " for "_"
x = list(df['Home State'])
x = [y.title().replace(" ", "_") for y in x ]
x
```

```python
# Saving the changes of the column value in a Series and passing its values to the "Home State" column of the Data Frame
df['Home State'] = pd.Series(x)
df
```

```python
del df['State Region']
```

```python
df.head()
```

```python
# For each cell of the column(standard axis=0) 'Home State', check the State Region and returns a string with this information
# Each return will be placed at the exact place where would be callen
df['Home State Region'] = df['Home State'].apply(return_region)
df
```
