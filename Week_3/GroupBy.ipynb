```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
for state in df['STNAME'].unique():
    avg = np.average(df[df['STNAME'] == state].dropna()['CENSUS2010POP'])
    
    print('State ' + state + ' has average 2010 population of ' + str(avg))
```
