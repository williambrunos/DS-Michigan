# Groupwise Analysis

In pandas, we can apply some kind of search for each group of the data frame with same caracheristics using ``groupby('column')``. This returns a data set of groups that has the fields 'points' with the same values.

``groupby()`` created a group of reviews which allotted the same point values to the given wines. Then, for each of these groups, we grabbed the points() column and counted how many times it appeared. value_counts() is just a shortcut to this groupby() operation.

```python
df.groupby('points').points.count() # This is exactly df['points'].value_counts()
```

Imagine we want to find the cheapest wine in the reviews for each group of points.

```python
'''
We are grouping the data in 'df' by the points, taking the values in 'price' column and taking the minimun value for each group
'''
df.groupby('points').price.min()
```

Using the ``groupby`` method, we slice the data frame in groups which each group has cells with matching values for the column we wanted to group.

So, for each subdataframe, we can apply functions to these subdatasets using ``apply(function)`` too.

```python
df.groupby('winery').apply(lambda df: df.title.iloc[0])
```

Imagine if we want to know the best wine for each country and province:

```python
df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

Another function that needs to be explored is the ``agg()`` function. This allow us to apply a list of funtions in a data frame or even in each cell of it.

Let's get some statistical values for the label price for each group of country on the data frame.

```python
df.groupby('country').price.agg([len, np.min, np.max, np.mean])
```

# Sorting

When grouping our data with ``groupby`` function, it will output for us a dataframe with values sorted by indixes, not by the value of the cells.

To sort the cells on thw way we want, we can use ``sort_values(by='column_name')``. One intereresting attribute to look for in this function is the 'ascending' one. This tells to our sort_values function how it will sort the values in the cell.

Eg: The standard value for ascending is True, which means it will sort the values in a ascending way. If we set it to False, sort_values will sort the cells in a descending way.

```python
df.groupby(['country', 'province']).price.agg([len, np.max, np.min, np.mean]).sort_values(by='len', ascending=False)
```

We can also sort the index of our data frame using ``sort_index()`` function

In addition, we can sort_values of the data frame by a list of columns doing: ``sort_values([list_of_columns])``.

```python
df.groupby(['country', 'province']).agg([len, np.max, np.min]).sort_index()
```
