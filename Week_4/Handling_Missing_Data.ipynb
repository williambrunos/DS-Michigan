# Missing Data

For a lot of reasons, sometimes a lot of data is missing on the data frame. These missing data are assigned as ``NaN(Not a Number)`` value of dtype as ``float64``.

```python
df = pd.read_csv('./datasets/winemag-data-130k-v2.csv', index_col=0)
df.head()
```

We can pass a boolean mask to see what celss has the ``NaN`` values.

```python
# We want to see the df where the cells from 'region_2' are NaN (or NULL)
df[pd.isnull(df.region_2)]
```

We have the ``fillna(new_value)`` function to fill the NaN cells with a specified value.

```python
# This will fill all the NaN cells from region_2 column with the 'Unknown' string
df.region_2.fillna('Unknown', inplace=True)
df
```

Alternatively, we may have a non-null value that we would like to replace. For example, suppose that since this dataset was published, reviewer Kerin O'Keefe has changed her Twitter handle from @kerinokeefe to @kerino. One way to reflect this in the dataset is using the replace() method:

```python
df.taster_twitter_handle.replace('@kerinokeefe', '@Kerino', inplace=True)
df
```

# Data Types (Dtype)

Dtype is the data type of a series of the Data Frame, or a column of it.

We can check this DType using ``dtype`` attribute for a column or all the whole data frame using the attribute ``dtypes`` with a 's'.

```python
df.price.dtype
```

```python
df.dtypes
```

We can use a kind of 'casting' for the values in a series using the ``astype('type')`` function. With this, we can create a series to see the values of the specified column as the type we passed as argument.

```python
df.price.astype('float64')
```
