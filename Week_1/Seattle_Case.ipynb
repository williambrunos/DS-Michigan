```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('./Seattle2014.csv')
df
```

```python
# The data set does not have 'trash' values (NaN)
df.isna().sum()
```

```python
plt.figure(figsize=(10, 8))
plt.hist(df['PRCP'], bins=50);
```

```python
df_cleaned_PRCP = df[df['PRCP'] > 0]
df_cleaned_PRCP
```

```python
plt.figure(figsize=(10, 8))
plt.title('Amount of days VS Rain preciptation')
plt.xlabel('Preciptation')
plt.ylabel('Amount of days')
plt.hist(df_cleaned_PRCP['PRCP'], bins=50);
```

```python
# How many rainy days were there that year (2014)?
df_cleaned_PRCP['PRCP'].describe()
# Response: 150 days
```

```python
df['SNOW'].value_counts()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Amount of days VS Rain preciptation')
plt.xlabel('Preciptation')
plt.ylabel('TMAX')
sns.lineplot(x=df_cleaned_PRCP['PRCP'], y=df_cleaned_PRCP['TMAX'], data=df_cleaned_PRCP);
```

```python
df_cleaned_PRCP['TMAX'].value_counts()
```

```python
sns.scatterplot(data=df_cleaned_PRCP, x='PRCP', y='TMIN');
```
